# merge() function

The merge() function in pandas is like a magic wand for combining data from different sources, kind of like merging puzzle pieces together. It takes two or more DataFrames and joins them based on a common column or index.

## Q. We have to find the orange cap holders (player with most runs in a season) of each seasons.

```python
import pandas as pd
delivery=pd.read_csv("deliveries.csv")
match=pd.read_csv("matches.csv")

new=delivery.merge(match,left_on="match_id",right_on="id")
new
```

```python
delivery.merge(match,left_on="match_id",right_on="id").shape
```

```python
new.groupby(["season","batsman"])["batsman_runs"].sum()
```

```python
type(new.groupby(["season","batsman"])["batsman_runs"].sum())           #Multi indexed series

```

## Note: reset_index() function is used to convert the series into dataframe.

```python
new.groupby(["season","batsman"])["batsman_runs"].sum().reset_index()
```

```python
new.groupby(["season","batsman"])["batsman_runs"].sum().sort_values(ascending=False).reset_index().drop_duplicates(subset="season",keep="first").sort_values("season")[["season","batsman"]]
```
